1- Leitura dos aqruivos e analise dos tipos de dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv',encoding='ANSI')
df.head()
```

```python
df.describe()
```

```python
print(df.isna().sum())

```

```python
df = df.dropna()
```

```python
filtro_01 = df['UnitPrice'].isna() | (df['UnitPrice'] < 0)

df_filtro_01 = df[filtro_01]

print(df_filtro_01)
```

```python
df = df[df['UnitPrice'] > 0]

print(df)
```

```python
filtro_quantity = df['Quantity'].isna() | (df['Quantity'] < 0)

df_filtro_quantity = df[filtro_quantity]

print(df_filtro_quantity)
```

```python
df = df[df['Quantity'] > 0]

print(df)
```

```python
linhas_duplicadas = df[df.duplicated()]
print(linhas_duplicadas)
```

```python
df = df.drop_duplicates()
```

```python
print(df.dtypes)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype('int64')
```

```python
print(df.dtypes)
```

```python
outliers_quantidade = df[df['Quantity'] > 10000]
outliers_preco = df[df['UnitPrice'] > 5000]
print(outliers_quantidade)
print(outliers_preco)
```

```python
df = df[df['Quantity'] <= 10000]
df = df[df['UnitPrice'] <= 5000]
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
ultima_compra = df['InvoiceDate'].max()

print("Data da última compra:", ultima_compra)
```

```python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
vendas_por_pais = df.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=vendas_por_pais.index, y=vendas_por_pais.values, palette='viridis')
plt.title('Top 10 Países com Maior Valor em Vendas')
plt.xlabel('País')
plt.ylabel('Valor Total em Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
quantidade_por_produto = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=quantidade_por_produto.index, y=quantidade_por_produto.values, palette='plasma')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.show()
```

```python
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')

vendas_por_mes = df.groupby('YearMonth')['TotalSales'].sum()

plt.figure(figsize=(10, 6))
sns.lineplot(x=vendas_por_mes.index.astype(str), y=vendas_por_mes.values, marker='o', color='b')
plt.title('Valor de Venda Total por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor Total em Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_paises = vendas_por_pais.head(10).index
df_top10_paises = df[df['Country'].isin(top_10_paises)]

vendas_por_mes_pais = df_top10_paises.groupby(['YearMonth', 'Country'])['TotalSales'].sum().unstack().fillna(0)

vendas_por_mes_pais.plot(figsize=(12, 8), marker='o', cmap='tab10')
plt.title('Valor de Venda Total por Mês e por País (Top 10 Países)')
plt.xlabel('Mês')
plt.ylabel('Valor Total em Vendas')
plt.xticks(rotation=45)
plt.legend(title='Países', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

ultima_compra = df['InvoiceDate'].max()

cliente_rfm = df.groupby(['CustomerID', 'InvoiceNo']).agg(
    compra_data=('InvoiceDate', 'max'), 
    valor_total=('TotalPrice', 'sum')     
).reset_index()

rfm = cliente_rfm.groupby('CustomerID').agg(
    recencia=('compra_data', lambda x: (ultima_compra - x.max()).days),  
    frequencia=('InvoiceNo', 'count'),                                   
    monetario=('valor_total', 'sum')                                     
).reset_index()

print(rfm.head())
```
